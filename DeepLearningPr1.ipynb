```
import IPython.display
from PIL import Image, ImageDraw
import numpy as np
from random import randint
 
# A simple function to display an image in an ipython notebook
def nbimage( data ):
    IPython.display.display(Image.fromarray(data))
    
def randomColor():
  r = randint(0,255)
  g = randint(0,255)
  b = randint(0,255)
  return "rgb(" + str(r) + "," + str(g) + "," + str(b) + ")"

# a 512x512 image, with 3 color channels (R,G,B)
data = np.zeros([512,512,3]) 
data = (255*data).astype('uint8')

im = Image.fromarray(data)
draw = ImageDraw.Draw(im)

# Draw 4 ellipses
for i in range(100):
  point = [randint(0,511),randint(0,511)]
  point.append(point[0] + randint(0,511-point[0]))
  point.append(point[1] + randint(0,511-point[1]))
  draw.ellipse(tuple(point), fill=randomColor())

del draw
 
# display it!
IPython.display.display(im)
```
